## EfficientNetB1

```python
import sys
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
import tensorflow as tf
import PIL.Image as Image
from tensorflow import keras
from tensorflow.keras.applications.imagenet_utils import decode_predictions
```

```python
model=tf.keras.applications.efficientnet.EfficientNetB1(
    include_top=True, 
    weights='imagenet', 
    input_tensor=None, 
    input_shape=None, 
    pooling=None, 
    classes=1000, 
    classifier_activation='softmax')
```

```python
img=Image.open(r'/content/B1-SAMPLE.jpg')
image=Image.open(r'/content/B1-SAMPLE.jpg')
```

```python
image
```

```python
img=cv2.resize(np.float32(img),(240,240))
```

```python
img.shape
```

```python
img
```

```python
img=img[np.newaxis,...]
```

```python
img.shape
```

```python
sampleprediction=model.predict(img)
```

```python
sampleprediction
```

```python
sampleans=np.argmax(sampleprediction)
```

```python
sampleans
```

```python
labels=pd.read_csv(r'/content/Label.txt',header=None)
```

```python
labels
```

```python
labels.iloc[sampleans]
```

```python
decode_predictions(sampleprediction,top=3)
```

```python
#checking the original image
plt.axis('off')
plt.imshow(image)
```

```python
print("Accuracy of prediction: ",round(decode_predictions(sampleprediction,top=1)[0][0][2]*100,2),"%")
```
